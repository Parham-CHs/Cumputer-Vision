# Requirements

```python
import cv2
import os
import matplotlib.pyplot as plt
```

# Loading Image

```python
image_address = os.path.join("images", "image.webp")

model = os.path.join("models", "haarcascade_frontalcatface_extended.xml")

image = cv2.imread(image_address)

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

This Python snippet prepares for image processing, likely for cat face detection.
It defines the file path for an input image ("image.webp") located in an "images" directory.
It also specifies the path to a Haar cascade model file ("haarcascade_frontalcatface_extended.xml") in a "models" directory.
The script then loads the specified image using OpenCV's `imread` function.
Finally, it converts the loaded image from the default BGR color space to RGB.

```python
plt.figure(figsize=(8, 8))

plt.imshow(image)

plt.axis("off")

plt.title("Original Image")
plt.show()

```

This Python code uses the Matplotlib library to display an image.
It first creates a new figure, setting its display size to 8x8 inches.
Then, `plt.imshow(image)` is used to render the image data stored in the `image` variable.
`plt.axis("off")` removes the x and y axes for a cleaner presentation.
Finally, `plt.title("Original Image")` adds a title, and `plt.show()` displays the image.

# Detecting Cat

```python
detector = cv2.CascadeClassifier(model)

rects = detector.detectMultiScale(image, scaleFactor=1.1, minNeighbors=5, minSize=(75, 75))

for (i, (x, y, w, h)) in enumerate(rects):
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cv2.putText(
        image,
        f"Cat #{i+1}",
        (x, y - 10),
        cv2.FONT_HERSHEY_SIMPLEX,
        0.8,
        (0, 255, 0),
        2
    )

```

This Python code snippet performs object detection using an OpenCV Haar Cascade Classifier.
It first loads the pre-trained cascade model specified by the `model` path.
Then, `detector.detectMultiScale` is called on the `image` to find regions that match the model, likely cat faces.
The code iterates through each detected object's coordinates (`rects`).
For each detection, it draws a green rectangle around it and labels it "Cat #[number]" on the `image`.

```python
plt.figure(figsize=(8, 8))

plt.imshow(image)

plt.axis("off")

plt.title("Cat Face Detection")
plt.show()
```

This Python code uses Matplotlib to display an image, presumably after cat faces have been detected and marked on it.
It creates a new figure with a specified size of 8x8 inches.
`plt.imshow(image)` then renders this (modified) image data.
`plt.axis("off")` is used to hide the x and y axes from the plot.
Finally, it sets the title to "Cat Face Detection" and `plt.show()` displays the image.
